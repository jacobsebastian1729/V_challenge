```python
import numpy as np
import pandas as pd
```

```python
pred_47 = pd.read_csv("prediction_47.csv")
pred_48 = pd.read_csv("prediction_48.csv")
pred_49 = pd.read_csv("prediction_49.csv")
pred_50 = pd.read_csv("prediction_50.csv")
pred_51 = pd.read_csv("prediction_51.csv")
pred_52 = pd.read_csv("prediction_52.csv")
pred_53 = pd.read_csv("prediction_53.csv")
```

```python
pred_47.columns = ['pred_47']
pred_48.columns = ['pred_48']
pred_49.columns = ['pred_49']
pred_50.columns = ['pred_50']
pred_51.columns = ['pred_51']
pred_52.columns = ['pred_52']
pred_53.columns = ['pred_53']

# Verify the changes (optional)
print(pred_47.head())
print(pred_48.head())
print(pred_49.head())
print(pred_50.head())
print(pred_51.head())
print(pred_52.head())
print(pred_53.head())
```

```python
concatenated_df = pd.concat([pred_47, pred_48, pred_49, pred_50, pred_51, pred_52, pred_53], axis=1)

# Verify the concatenated DataFrame
print(concatenated_df.head())
```

```python
mapping = {'Low': 1, 'Medium': 2, 'High': 3}
for col in concatenated_df.columns:
    concatenated_df[col] = concatenated_df[col].map(mapping)

# Calculate the weighted average (assuming equal weight)
concatenated_df['pred'] = concatenated_df.mean(axis=1)

# Map the numerical values back to 'Low', 'Medium', 'High'
reverse_mapping = {1: 'Low', 2: 'Medium', 3: 'High'}
concatenated_df['pred'] = concatenated_df['pred'].round().map(reverse_mapping)

```

```python
prediction = concatenated_df[['pred']].copy()
```

```python
prediction.head()
```

```python
import os
import zipfile

# Save the DataFrame to a CSV file
prediction.to_csv('prediction.csv', index=False)

def compress_file(input_file, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(input_file, os.path.basename(input_file))
input_file = 'prediction.csv'  # Input file to compress
output_zip = 'prediction.csv.zip'  # Output ZIP archive
compress_file(input_file, output_zip)
```
