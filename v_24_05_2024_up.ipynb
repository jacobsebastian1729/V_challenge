# **logistic regression combined**

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("prediction_1.csv")
df2 = pd.read_csv("prediction_2.csv")
df3 = pd.read_csv("prediction_3.csv")
df4 = pd.read_csv("prediction_4.csv")
df5 = pd.read_csv("prediction_5.csv")
```

```python
df1 = df1.rename(columns={'pred': 'pred_1'})
df2 = df2.rename(columns={'pred': 'pred_2'})
df3 = df3.rename(columns={'pred': 'pred_3'})
df4 = df4.rename(columns={'pred': 'pred_4'})
df5 = df5.rename(columns={'pred': 'pred_5'})
```

```python
df5.head()
```

```python
combined_df = pd.concat([df1, df2, df3, df4, df5], axis=1)
```

```python
combined_prediction = combined_df.mode(axis=1)[0]

# Create a new DataFrame with the combined predictions
output_df = pd.DataFrame({'pred': combined_prediction})
```

```python
combined_df['combined_prediction'] = combined_prediction
```

```python
combined_df.head()
```

```python
matches = combined_df.iloc[:, :-1].eq(combined_df['combined_prediction'], axis=0)

# Add the combined prediction column to the matches DataFrame for clarity
matches['combined_prediction'] = combined_df['combined_prediction']

```

```python
matches.head()
```

```python
match_rates = matches.iloc[:,:-1].mean() * 100

# Print the matches DataFrame
print(match_rates)
```

```python
output_df.to_csv('prediction.csv', index=False)
import os
import zipfile
def compress_file(input_file, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(input_file, os.path.basename(input_file))
input_file = 'prediction.csv'  # Input file to compress
output_zip = 'prediction.csv.zip'  # Output ZIP archive
compress_file(input_file, output_zip)
```
