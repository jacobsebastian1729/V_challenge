```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_merged_4.csv")
```

```python
#df.head()
```

```python
null_columns = df.columns[df.isnull().all()]
```

```python
df2 = df.drop(columns=null_columns)
```

```python
#df2.head()
```

```python
#print(df2.columns[17:25])
```

```python
selected_columns = df2[['risk_level', df2.columns[14]]]
print(selected_columns.head(2))
print(min(selected_columns.iloc[:, 1]))
print(max(selected_columns.iloc[:, 1]))
```

```python
high_risk = selected_columns[selected_columns['risk_level'] == 'High']
medium_risk = selected_columns[selected_columns['risk_level'] == 'Medium']
low_risk = selected_columns[selected_columns['risk_level'] == 'Low']

# Determine the common range for all histograms
min_value = min(selected_columns.iloc[:, 1])
max_value = max(selected_columns.iloc[:, 1])

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

# Plot histograms for each risk level with the same x-axis range
for ax, data, label, color in zip(axs, [high_risk, medium_risk, low_risk], ['High Risk', 'Medium Risk', 'Low Risk'], ['red', 'blue', 'green']):
    counts, bins, _ = ax.hist(data.iloc[:, 1], color=color, alpha=0.7, range=(min_value, max_value), bins=300)  # Adjust the number of bins as needed
    ax.set_title(label)
    ax.set_ylabel('Frequency')

    # Add percentage annotations
    for count, bin in zip(counts, bins):
        percent = count / len(data) * 100
        #ax.text(bin + (bins[1] - bins[0]) / 2, count + 0.5, f'{percent:.1f}%', ha='center', va='bottom', fontsize=8)

axs[-1].set_xlabel('Values')

plt.tight_layout()
plt.show()
```
